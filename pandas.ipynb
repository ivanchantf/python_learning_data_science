```python
mean=flight.duration.mean()
flight.duration.map(lambda c: c-mean) #no change on flight
```

```python
#Faster
mean=flight.duration.mean()
flight.duration-mean
```

```python
#Alternatives :using apply ()
mean=flight.duration.mean()
def duration(row):
    row.duration = row.duration - mean
    return row
flight.apply(duration,axis='columns')
```

# Grouping & Sorting

```python
flight.groupby('airline')
flight.groupby('airline').airline.count() #same as flight.airline.value_count()
```

```python
flight.groupby('airline').price.mean()
```

```python
flight.groupby('airline').apply(lambda df: df.loc[df.price.idxmax()])
```

```python
flight.groupby('airline').price.agg([len,max,min]) #len is count here
```

```python
flight.groupby('airline').price.agg([len,max,min]).sort_values(by='max',ascending=False)
```

```python
flight.sort_values(by=['airline','flight'],ascending=False)
```

```python
flight.groupby('airline').price.max().sort_index()
```

```python
flight.groupby(['airline','flight']).agg([len]).price.sort_values(by='len',ascending=False)
```

```python
flight.groupby(['airline','flight']).size().sort_values(ascending=False) #use the function size() instead of agg['len']
```
