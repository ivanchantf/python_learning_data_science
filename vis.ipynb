# Visualizing Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv('weather.csv',index_col="time",parse_dates=True)
weather.head()
```

## sns.lineplot

```python
weather2= weather.loc['2000-01-01':'2000-01-31',['relative_humidity']]
print(weather2.count())
plt.figure(figsize=(20,15))
plt.title('hum over time')
sns.lineplot(data=weather2['relative_humidity'],label='hummm')
plt.xlabel('timeline')
plt.ylabel('hum')
```

## sns.barplot(x,y)


```python
ign_data= pd.read_csv('ign_scores.csv',index_col='Platform')
ign_data
```

```python
plt.figure(figsize=(30,12))
sns.barplot(x=ign_data.index,y=ign_data['Action'])
plt.ylabel='Action game score'
plt.xlabel='platform'
plt.title('Scores of Action games on different platform')
```

## sns.heatmap(data=, annot)

```python
sns.heatmap(data=ign_data,annot=True)
```

## sns.scatterplot(x=,y=)

```python
candy = pd.read_csv('candy.csv')
```

```python
sns.scatterplot(x=candy['sugarpercent'],y=candy['pricepercent'])
```

## sns.regplot(x=,y=)

```python
sns.regplot(x=candy['sugarpercent'],y=candy['pricepercent'])
```

## sns.scatterplot(x=,y=) With Hue

```python
sns.scatterplot(x=candy['pricepercent'],y=candy['winpercent'],hue=candy['chocolate'])
```

## sns.lmplot(x=,y=,hue=,data=)

```python
sns.lmplot(x='pricepercent',y='winpercent',hue='chocolate',data=candy)
```

## sns.swarmplot(x=,y=)

```python
sns.swarmplot(x=candy['chocolate'],y=candy['winpercent'])
```
