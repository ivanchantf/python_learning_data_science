# Visualizing Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv('Weather_dataset.csv',index_col="time",parse_dates=True)
weather.head()
```

## sns.lineplot

```python
weather2= weather.loc['2000-01-01':'2000-01-31',['relative_humidity']]
print(weather2.count())
plt.figure(figsize=(20,15))
plt.title('hum over time')
sns.lineplot(data=weather2['relative_humidity'],label='hummm')
plt.xlabel('timeline')
plt.ylabel('hum')
```
